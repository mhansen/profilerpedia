```python
import pandas as pd
```

```python
def read_sheet(sheet, **kwargs):
    return pd.read_csv(f"https://docs.google.com/spreadsheets/d/1cVcHofphkQqk1yGeuBPVTit8HQ0oa5SlRM6gkHIagtw/gviz/tq?tqx=out:csv&sheet={sheet}", **kwargs)
```

```python
profilers = read_sheet("Profilers", index_col="Profiler", na_filter=False).sort_index(key=lambda x: x.str.lower())
profilers["Star"] = profilers["⭐"]
profilers
```

```python
outputs = read_sheet("Outputs", na_filter=False)
outputs
```

```python
formats = read_sheet("Formats", index_col="Name", na_filter=False).sort_index(key=lambda x: x.str.lower())
formats["Star"] = formats["⭐"]
formats
```

```python
converters = read_sheet("Converters", na_filter=False)
converters
```

```python
visualises = read_sheet("Visualises", na_filter=False)
visualises
```

```python
uis = read_sheet("UIs", index_col="UI", na_filter=False).sort_index(key=lambda x: x.str.lower())
uis["Star"] = uis["⭐"]
uis
```

```python
import os
import graphviz
import json
from tqdm import tqdm


from jinja2 import Environment, FileSystemLoader, select_autoescape
env = Environment(
    loader=FileSystemLoader(searchpath="."),
    autoescape=select_autoescape(),
)

args = {
    "profilers": list(profilers.itertuples()),
    "outputs_by_format": {k: list(v.sort_values(by="Profiler", key=lambda x: x.str.lower()).itertuples()) for k, v in outputs.groupby("Format")},
    "outputs_by_profiler": {k: list(v.sort_values(by="Format", key=lambda x: x.str.lower()).itertuples()) for k, v in outputs.groupby("Profiler")},
    "formats": list(formats.itertuples()),
    "converters_by_to": {k: list(v.sort_values(by="From", key=lambda x: x.str.lower()).itertuples()) for k, v in converters.groupby("To")},
    "converters_by_from": {k: list(v.sort_values(by="To", key=lambda x: x.str.lower()).itertuples()) for k, v in converters.groupby("From")},
    "visualises_by_format": {k: list(v.sort_values(by="UI", key=lambda x: x.str.lower()).itertuples()) for k, v in visualises.groupby("Format")},
    "visualises_by_ui": {k: list(v.sort_values(by="Format", key=lambda x: x.str.lower()).itertuples()) for k, v in visualises.groupby("UI")},
    "uis": list(uis.itertuples()),
    # Maybe one day we can do without these converters, by using the slugs as IDs
    "uis_by_name": {ui.Index: ui for ui in uis.itertuples()},
    "profilers_by_name": {p.Index: p for p in profilers.itertuples()},
    "formats_by_name": {f.Index: f for f in formats.itertuples()},
    "converters_by_name": {c.Converter: c for c in converters.itertuples()},
}

def add_format_node(dot, format):
    node_id = f"format/{format.Slug}"
    dot.node(
        name=node_id,
        label=format.Index,
        shape="cylinder",
        href=f"/formats/{format.Slug}/",
        fontcolor="blue",
        target="_top",
    )
    return node_id

def add_profiler_node(dot, profiler):
    node_id = f"profiler/{profiler.Slug}"
    dot.node(
        name=node_id,
        label=profiler.Index,
        href=f"/profilers/{profiler.Slug}/",
        fontcolor="blue",
        target="_top",
        shape="box",
    )
    return node_id

def add_ui_node(dot, ui):
    node_id = f"ui/{ui.Slug}"
    dot.node(
        name=node_id,
        label=ui.Index,
        shape="folder",
        href=f"/uis/{ui.Slug}/",
        fontcolor="blue",
        target="_top",
    )
    return node_id
```

```python
graph_attrs = {
    "fontname": "Open Sans",
    "rankdir": "LR",
}

def format_dot(format):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    format_node_id = add_format_node(dot, format)
    outputs = args["outputs_by_format"].get(format.Index, [])
    for output in outputs:
        profiler = args["profilers_by_name"][output.Profiler]
        dot.edge(
            add_profiler_node(dot, profiler),
            format_node_id,
        )

    visualises = args["visualises_by_format"].get(format.Index, [])
    for visualise in visualises:
        ui = args["uis_by_name"][visualise.UI]
        dot.edge(
            format_node_id,
            add_ui_node(dot, ui),
        )
    return dot

for format in tqdm(formats.itertuples(), total=formats.shape[0]):
    base = os.path.realpath("content/en/formats/")
    path = os.path.realpath(base + "/" + format.Slug)
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    with open(path + ".html", "w") as f:
        f.write(env.get_template("format.html.j2").render(
            format=format,
            **args,
        ))
    format_dot(format).render(directory=base, filename=path + ".dot", quiet=False, format='svg')
```

```python
def profiler_dot(profiler):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    # profiler_cluster = graphviz.Digraph(name='cluster_profiler', graph_attr={"label":"Profiler 🩺"})
    profiler_node_id = add_profiler_node(dot, profiler)
    # dot.subgraph(profiler_cluster)
    outputs = args["outputs_by_profiler"].get(profiler.Index, [])
    for output in outputs:
        format = args["formats_by_name"][output.Format]
        format_node_id = add_format_node(dot, format)
        dot.edge(
            profiler_node_id,
            format_node_id,
        )
        visualises = args["visualises_by_format"].get(format.Index, [])
        for visualise in visualises:
            ui = args["uis_by_name"][visualise.UI]
            dot.edge(
                format_node_id,
                add_ui_node(dot, ui),
            )
    return dot

for profiler in tqdm(profilers.itertuples(), total=profilers.shape[0]):
    base = os.path.realpath("content/en/profilers/")
    path = os.path.realpath(base + "/" + profiler.Slug)
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    with open(path + ".html", "w") as f:
        json.dump(
            {
                "title": profiler.Index + " " + profiler.Star,
                "platform": profiler.Platform,
                "docs": profiler.Docs,
                "slug": profiler.Slug,
                "notes": profiler.Notes,
                "formats": [
                    {
                        "format": o.Format,
                        "slug": args["formats_by_name"][o.Format].Slug,
                        "docs": o.Docs,
                    }
                    for o in args["outputs_by_profiler"].get(profiler.Index, [])],
                "profiler_platforms": profiler.Platform.split(","),
            },
            fp=f,
            indent=2,
            sort_keys=True
        )
        # Need a newline
        f.write("\n")
    # profiler_dot(profiler).render(directory=base, filename=path + ".dot", quiet=False, format='svg')
```

```python
def converter_dot(converter):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    format_from = args["formats_by_name"][converter.From]
    format_to   = args["formats_by_name"][converter.To]
    dot.edge(
        add_format_node(dot, format_from),
        add_format_node(dot, format_to),
        label=converter.Converter,
    )
    return dot
    
for converter in tqdm(converters.itertuples(), total=converters.shape[0]):
    base = os.path.realpath("content/en/converters/")
    path = os.path.realpath(base + "/" + converter.Slug)
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    with open(path + ".html", "w") as f:
        f.write(env.get_template("converter.html.j2").render(
            converter=converter,
            **args,
        ))
    converter_dot(converter).render(directory=base, filename=path + ".dot", quiet=False, format='svg')
```

```python
def ui_dot(ui):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    ui_node_id = add_ui_node(dot, ui)
    for visualise in args["visualises_by_ui"].get(ui.Index, []):
        format = args["formats_by_name"][visualise.Format]
        format_node_id = add_format_node(dot, format)
        dot.edge(
            format_node_id,
            ui_node_id,
        )
        for output in args["outputs_by_format"].get(format.Index, []):
            profiler = args["profilers_by_name"][output.Profiler]
            dot.edge(
                add_profiler_node(dot, profiler),
                format_node_id
            )
    return dot

for ui in tqdm(uis.itertuples(), total=uis.shape[0]):
    base = os.path.realpath("content/en/uis/")
    path = os.path.realpath(base + "/" + ui.Slug)
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    with open(path + ".html", "w") as f:
        f.write(env.get_template("ui.html.j2").render(
            ui=ui,
            **args,
        ))
    ui_dot(ui).render(directory=base, filename=path + ".dot", quiet=False, format='svg')
```
