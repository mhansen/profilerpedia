```python
import pandas as pd
```

```python
def read_sheet(sheet, **kwargs):
    return pd.read_csv(f"https://docs.google.com/spreadsheets/d/1cVcHofphkQqk1yGeuBPVTit8HQ0oa5SlRM6gkHIagtw/gviz/tq?tqx=out:csv&sheet={sheet}", **kwargs)
```

```python
profilers = read_sheet("Profilers", index_col="Profiler", na_filter=False).sort_index(key=lambda x: x.str.lower())
profilers["Star"] = profilers["‚≠ê"]
profilers
```

```python
outputs = read_sheet("Outputs", na_filter=False)
outputs
```

```python
formats = read_sheet("Formats", index_col="Name", na_filter=False).sort_index(key=lambda x: x.str.lower())
formats["Star"] = formats["‚≠ê"]
formats
```

```python
converters = read_sheet("Converters", na_filter=False)
converters
```

```python
visualises = read_sheet("Visualises", na_filter=False)
visualises
```

```python
uis = read_sheet("UIs", index_col="UI", na_filter=False).sort_index(key=lambda x: x.str.lower())
uis["Star"] = uis["‚≠ê"]
uis
```

```python
import os
import graphviz
import json
from tqdm.notebook import tqdm

args = {
    "profilers": list(profilers.itertuples()),
    "outputs_by_format": {k: list(v.sort_values(by="Profiler", key=lambda x: x.str.lower()).itertuples()) for k, v in outputs.groupby("Format")},
    "outputs_by_profiler": {k: list(v.sort_values(by="Format", key=lambda x: x.str.lower()).itertuples()) for k, v in outputs.groupby("Profiler")},
    "formats": list(formats.itertuples()),
    "converters_by_to": {k: list(v.sort_values(by="From", key=lambda x: x.str.lower()).itertuples()) for k, v in converters.groupby("To")},
    "converters_by_from": {k: list(v.sort_values(by="To", key=lambda x: x.str.lower()).itertuples()) for k, v in converters.groupby("From")},
    "visualises_by_format": {k: list(v.sort_values(by="UI", key=lambda x: x.str.lower()).itertuples()) for k, v in visualises.groupby("Format")},
    "visualises_by_ui": {k: list(v.sort_values(by="Format", key=lambda x: x.str.lower()).itertuples()) for k, v in visualises.groupby("UI")},
    "uis": list(uis.itertuples()),
    # Maybe one day we can do without these converters, by using the slugs as IDs
    "uis_by_name": {ui.Index: ui for ui in uis.itertuples()},
    "profilers_by_name": {p.Index: p for p in profilers.itertuples()},
    "formats_by_name": {f.Index: f for f in formats.itertuples()},
    "converters_by_name": {c.Converter: c for c in converters.itertuples()},
}

def format_node_id(format):
    return f"format/{format.Slug}"

def add_format_node(dot, format, **kwargs):
    node_id = format_node_id(format)
    dot.node(
        name=node_id,
        label=format.Index,
        shape="cylinder",
        href=f"/formats/{format.Slug}/",
        fontcolor="blue",
        target="_top",
        **kwargs,
    )
    return node_id

def profiler_node_id(profiler):
    return f"profiler/{profiler.Slug}"


def add_profiler_node(dot, profiler, **kwargs):
    node_id = profiler_node_id(profiler)
    dot.node(
        name=node_id,
        label=profiler.Index,
        href=f"/profilers/{profiler.Slug}/",
        fontcolor="blue",
        target="_top",
        shape="box",
        **kwargs,
    )
    return node_id

def add_ui_node(dot, ui, **kwargs):
    node_id = f"ui/{ui.Slug}"
    dot.node(
        name=node_id,
        label=ui.Index,
        shape="folder",
        href=f"/uis/{ui.Slug}/",
        fontcolor="blue",
        target="_top",
        **kwargs,
    )
    return node_id

graph_attrs = {
    "fontname": "Open Sans",
    "rankdir": "LR",
    "stylesheet": "/graphviz.css",
    # Allows styling the background color from CSS.
    "bgcolor": "transparent",
}
```

```python

def format_dot(format):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    format_node_id = add_format_node(dot, format, **{"class": "current"})
    outputs = args["outputs_by_format"].get(format.Index, [])
    for output in outputs:
        profiler = args["profilers_by_name"][output.Profiler]
        dot.edge(
            add_profiler_node(dot, profiler),
            format_node_id,
        )

    visualises = args["visualises_by_format"].get(format.Index, [])
    for visualise in visualises:
        ui = args["uis_by_name"][visualise.UI]
        dot.edge(
            format_node_id,
            add_ui_node(dot, ui),
        )
    return dot

for format in tqdm(formats.itertuples(), total=formats.shape[0]):
    base = os.path.realpath("content/en/formats/")
    if not format.Slug:
        continue
    path = os.path.realpath(base + "/" + format.Slug)
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    with open(path + ".html", "w") as f:
        json.dump(
            {
                "title": format.Index + " " + format.Star,
                "star": format.Star,
                "project_site": format.ProjectSite,
                "schema": format.Schema,
                "slug": format.Slug,
                "notes": format.Notes,
                "file_format": format.FileFormat,
                "timestamping": format.Timestamping,
                # Don't conflict with hugo "outputs" key
                "outputs_": [
                  {
                      "docs": o.Docs,
                      "profiler": o.Profiler,
                      "profiler_slug": args["profilers_by_name"][o.Profiler].Slug,
                  }
                  for o in args["outputs_by_format"].get(format.Index, [])
                ],
                "visualises": [
                  {
                      "docs": v.Docs,
                      "ui": v.UI,
                      "ui_slug": args["uis_by_name"][v.UI].Slug,
                  }
                  for v in args["visualises_by_format"].get(format.Index, [])
                ],
                "convertable_from": [
                    {
                        "docs": c.Docs,
                        "format": c.From,
                        "format_slug": args["formats_by_name"][c.From].Slug,
                        "converter": c.Converter,
                        "converter_slug": c.Slug,
                    }
                    for c in args["converters_by_to"].get(format.Index, [])
                ],
                "convertable_to": [
                    {
                        "docs": c.Docs,
                        "format": c.To,
                        "format_slug": args["formats_by_name"][c.To].Slug,
                        "converter": c.Converter,
                        "converter_slug": c.Slug,
                    }
                    for c in args["converters_by_from"].get(format.Index, [])
                ],
            },
            fp=f,
            indent=2,
            sort_keys=True
        )
        # Need a newline
        f.write("\n")
    format_dot(format).render(directory=base, filename=path + ".dot", quiet=False, format='svg')
```

```python
def profiler_dot(profiler):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    profiler_node_id = add_profiler_node(dot, profiler, **{"class": "current"})
    outputs = args["outputs_by_profiler"].get(profiler.Index, [])
    for output in outputs:
        format = args["formats_by_name"][output.Format]
        format_node_id = add_format_node(dot, format)
        dot.edge(
            profiler_node_id,
            format_node_id,
        )
        visualises = args["visualises_by_format"].get(format.Index, [])
        for visualise in visualises:
            ui = args["uis_by_name"][visualise.UI]
            dot.edge(
                format_node_id,
                add_ui_node(dot, ui),
            )
    return dot

for profiler in tqdm(profilers.itertuples(), total=profilers.shape[0]):
    base = os.path.realpath("content/en/profilers/")
    if not profiler.Slug:
        continue
    path = os.path.realpath(base + "/" + profiler.Slug)
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    with open(path + ".html", "w") as f:
        json.dump(
            {
                "title": profiler.Index + " " + profiler.Star,
                "star": profiler.Star,
                "platform": profiler.Platform,
                "docs": profiler.Docs,
                "slug": profiler.Slug,
                "notes": profiler.Notes,
                "formats": [
                    {
                        "format": o.Format,
                        "slug": args["formats_by_name"][o.Format].Slug,
                        "docs": o.Docs,
                    }
                    for o in args["outputs_by_profiler"].get(profiler.Index, [])],
                "profiler_platforms": [p for p in profiler.Platform.split(",") if p],
            },
            fp=f,
            indent=2,
            sort_keys=True
        )
        # Need a newline
        f.write("\n")
    profiler_dot(profiler).render(directory=base, filename=path + ".dot", quiet=False, format='svg')
```

```python
def converter_dot(converter):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    format_from = args["formats_by_name"][converter.From]
    format_to   = args["formats_by_name"][converter.To]
    dot.edge(
        add_format_node(dot, format_from),
        add_format_node(dot, format_to),
        label=converter.Converter,
        **{"class": "current"},
    )
    return dot
    
for converter in tqdm(converters.itertuples(), total=converters.shape[0]):
    base = os.path.realpath("content/en/converters/")
    path = os.path.realpath(base + "/" + converter.Slug)
    if not converter.Slug:
        continue
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    with open(path + ".html", "w") as f:
        json.dump(
            {
                "title": converter.Converter,
                "docs": converter.Docs,
                "note": converter.Note,
                "converts_from": {
                    "slug": args["formats_by_name"][converter.From].Slug,
                    "name": converter.From,
                },
                "converts_to": {
                    "slug": args["formats_by_name"][converter.To].Slug,
                    "name": converter.To,
                },
                "slug": converter.Slug,
            },
            fp=f,
            indent=2,
            sort_keys=True
        )
        # Need a newline
        f.write("\n")
    converter_dot(converter).render(directory=base, filename=path + ".dot", quiet=False, format='svg')
```

```python
def ui_dot(ui):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    ui_node_id = add_ui_node(dot, ui, **{"class": "current"})
    for visualise in args["visualises_by_ui"].get(ui.Index, []):
        format = args["formats_by_name"][visualise.Format]
        format_node_id = add_format_node(dot, format)
        dot.edge(
            format_node_id,
            ui_node_id,
        )
        for output in args["outputs_by_format"].get(format.Index, []):
            profiler = args["profilers_by_name"][output.Profiler]
            dot.edge(
                add_profiler_node(dot, profiler),
                format_node_id
            )
    return dot

for ui in tqdm(uis.itertuples(), total=uis.shape[0]):
    base = os.path.realpath("content/en/uis/")
    if not ui.Slug:
        continue
    path = os.path.realpath(base + "/" + ui.Slug)
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    with open(path + ".html", "w") as f:
        json.dump(
            {
                "title": ui.Index + " " + ui.Star,
                "star": ui.Star,
                "platform": ui.Platform,
                "site": ui.Site,
                "slug": ui.Slug,
                "notes": ui.Notes,
                "ui_platforms": [p for p in ui.Platform.split(",") if p],
                "visualises": [
                    {
                        "format": v.Format,
                        "format_slug": args["formats_by_name"][v.Format].Slug,
                        "docs": v.Docs,
                    }
                    for v in args["visualises_by_ui"].get(ui.Index, [])]
            },
            fp=f,
            indent=2,
            sort_keys=True
        )
        # Need a newline
        f.write("\n")
    ui_dot(ui).render(directory=base, filename=path + ".dot", quiet=False, format='svg')
```

```python
# search!
# output all the conversions
# todo: should I avoid loops? I think so.
# folded stacks and pprof are notorious for loops. makes the graphs hard to understand i reckon.
# the trouble is: it's cool and good to have multiple paths to an item.
# should we always choose the shortest path to avoid loops?
# I think disallowing backlinks is one good way to do this.
# possibly I should also split the graph in two:
# - one graph that's for forwards
# - one graph that's for backwards
# then they'll be disjoint.
def search_backwards(dot, format, seen_nodes, seen_conversions, nodes_from_root):
    print("searching backwards ", [f.Slug for f in nodes_from_root], format.Slug)
    if format not in seen_nodes:
        add_format_node(dot, format)
        seen_nodes.add(format)

    for c in args["converters_by_to"].get(format.Index, []):
        if c in seen_conversions:
            continue
        seen_conversions.add(c)
        from_format = args["formats_by_name"][c.From]
        if from_format in nodes_from_root:
            print("seen in parents already, returning")
            return
        dot.edge(
            format_node_id(from_format),
            format_node_id(format),
            label=c.Converter,
            href=f"/converters/{c.Slug}/",
            fontcolor="blue",
            target="_top",
        )
        search_backwards(dot, from_format, seen_nodes, seen_conversions, nodes_from_root + [format])

def search_forwards(dot, format, seen_nodes, seen_conversions, nodes_from_root):
    print("searching forwards ", [f.Slug for f in nodes_from_root], format.Slug)
    if format not in seen_nodes:
        add_format_node(dot, format)
        seen_nodes.add(format)

    for c in args["converters_by_from"].get(format.Index, []):
        if c in seen_conversions:
            continue
        seen_conversions.add(c)
        to_format = args["formats_by_name"][c.To]
        if to_format in nodes_from_root:
            print("seen in parents already, returning")
            return
        dot.edge(
            format_node_id(format),
            format_node_id(to_format),
            label=c.Converter,
            href=f"/converters/{c.Slug}/",
            fontcolor="blue",
            target="_top",
        )
        search_forwards(dot, to_format, seen_nodes, seen_conversions, nodes_from_root + [format])

    
def format_conversions_dot(format):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    add_format_node(dot, format, **{"class": "current"})
    seen_nodes = set(format)
    seen_conversions = set()
    nodes_from_root = []
    search_backwards(dot, format, seen_nodes, seen_conversions, nodes_from_root)
    nodes_from_root = []
    search_forwards(dot, format, seen_nodes, seen_conversions, nodes_from_root)
    return dot

def format_conversions_backwards_dot(format):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    add_format_node(dot, format, **{"class": "current"})
    seen_nodes = set(format)
    seen_conversions = set()
    nodes_from_root = []
    search_backwards(dot, format, seen_nodes, seen_conversions, nodes_from_root)
    return dot

def format_conversions_forwards_dot(format):
    dot = graphviz.Digraph(graph_attr=graph_attrs)
    add_format_node(dot, format, **{"class": "current"})
    seen_nodes = set(format)
    seen_conversions = set()
    nodes_from_root = []
    search_forwards(dot, format, seen_nodes, seen_conversions, nodes_from_root)
    return dot

for format in tqdm(formats.itertuples(), total=formats.shape[0]):
# for format in formats[formats['Slug'] == 'pprof'].itertuples():
    base = os.path.realpath("content/en/formats/")
    if not format.Slug:
        continue
    path = os.path.realpath(base + "/" + format.Slug)
    # Avoid path traversal
    commonpath = os.path.commonprefix((path, base))
    if commonpath != base:
        raise Exception("Bad path %s: common path=%s" % (path, commonpath))
    format_conversions_dot(format).render(
        directory=base,
        filename=path + ".converters.dot",
        quiet=False,
        format='svg',
    )
    format_conversions_backwards_dot(format).render(
        directory=base,
        filename=path + ".backwards.dot",
        quiet=False,
        format='svg',
    )
    format_conversions_forwards_dot(format).render(
        directory=base,
        filename=path + ".forwards.dot",
        quiet=False,
        format='svg',
    )
    
```
